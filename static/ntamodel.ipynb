```python
df.columns = [
'src_ip','src_port','dst_ip','dst_port','proto','state','duration',
'sbytes','dbytes','sttl','dttl','sloss','dloss',
'service','sload','dload','spkts','dpkts',
'swin','dwin','stcpb','dtcpb','smean','dmean',
'trans_depth','res_len','sjit','djit','stime','ltime',
'sintpkt','dintpkt','tcprtt','synack','ackdat',
'sm_ips_ports','state_ttl','http_method',
'ftp_login','ftp_cmd','srv_src','srv_dst',
'dst_ltm','src_ltm','src_dport_ltm',
'dst_sport_ltm','dst_src_ltm','attack_cat','label'
]
```

```python
df = df.drop('attack_cat', axis=1)
```

```python
df.info()
```

```python
df = df.drop(['src_ip', 'dst_ip'], axis=1)
```

```python
import numpy as np
def convert_port(x):
    try:
        x = str(x).strip()
        if x == '-' or x == '':
            return np.nan 
        if x.startswith('0x'):
            return int(x, 16)
        return int(x)
    except:
        return np.nan   
df['src_port'] = df['src_port'].apply(convert_port)
df['dst_port'] = df['dst_port'].apply(convert_port)
```

```python
df[['src_port','dst_port']].isna().sum()
```

```python
df.dropna(subset=['src_port', 'dst_port'], inplace=True)
```

```python
df['src_port'] = df['src_port'].astype(int)
df['dst_port'] = df['dst_port'].astype(int)
```

```python
df[['src_port','dst_port']].isna().sum()
df.shape
```

```python
df.columns
```

```python
df['label'].value_counts()
```

```python
df = pd.get_dummies(df, columns=['proto', 'state', 'service'], drop_first=True)
```

```python
x = df.drop('label', axis=1)
y = df['label']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state=42,stratify=y)
```

```python
model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
y_prob = model.predict_proba(x_test)[:,1]
y_pred_custom = (y_prob > 0.4).astype(int)
```

```python
print(confusion_matrix(y_test, y_pred_custom))
print(classification_report(y_test, y_pred_custom))
```
